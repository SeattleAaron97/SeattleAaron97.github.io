```python
import pandas as pd
import matplotlib.pyplot as plt

data_frame = pd.read_csv('C:/Users/aaron/Code/Data/Rental_Property_Registration.csv')

pd.set_option('display.max_columns', None) ### This line makes all the columns display, rather than ellipses shorten

data_frame.head()
```

```python
x = data_frame['Longitude']
y = data_frame['Latitude']

plt.hexbin(x, y, gridsize = 50, cmap = 'Greens')
plt.title('Seattle Rental Properties')
plt.rcParams["figure.figsize"] = (5,8.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
#### Make new columns for groups of lat/long

data_frame['short_lat'] = data_frame.Latitude.round(3)
data_frame['short_long'] = data_frame.Longitude.round(3)
```

```python
x = data_frame['short_long']
y = data_frame['short_lat']

plt.hexbin(x, y, gridsize = 50, cmap = 'Greens')
plt.title('Seattle Rental Properties')
plt.rcParams["figure.figsize"] = (5,8.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
