```python
import pandas as pd
import matplotlib.pyplot as plt

data_frame = pd.read_csv('C:/Users/aaron/Code/Data/Oregon_Census_Tracts.csv')

pd.set_option('display.max_columns', None) ### This line makes all the columns display, rather than ellipses shorten

data_frame.head()
```

```python
data_frame.info()
```

```python
print('POP20 =', data_frame['POP20'].sum())
print('POP20_LT18 =', data_frame['P20_LT18'].sum())
print('POP20_GT18 =', data_frame['P20_GE18'].sum())
```

```python
data_frame.plot(x = 'INTPTLON', xlabel = 'Longitude', y = 'INTPTLAT', ylabel = 'Latitude', kind = 'scatter', s = 0.2, title = 'Oregon by pop tract location')
```

```python
### Ok now I am going to make two new columns to shorten latitude and longtitude so i can categorize them

#data_frame['short_long'] = data_frame.round({'INTPTLON':2})
#data_frame['short_lat'] = data_frame.round({'INTPTLAT':2})

data_frame['short_lat'] = data_frame.INTPTLAT.round(1)
data_frame['short_long'] = data_frame.INTPTLON.round(1)


data_frame.head()
```

```python
data_frame.plot(x = 'short_long', xlabel = 'Longitude', y = 'short_lat', ylabel = 'Latitude', kind = 'scatter', s = 2, title = 'Oregon by pop tract location')
plt.rcParams["figure.figsize"] = (10,5.5)
```

```python
long_pop_frame = data_frame.groupby('short_long').agg({'POP20':'sum'}).reset_index()

long_pop_frame
```

```python
long_pop_frame.plot(title = 'Population Map by Longitude', kind = 'bar', figsize = (15,10), legend = 'Population')
```

```python
longitude = long_pop_frame['short_long']
pop = long_pop_frame['POP20']

fig, ax = plt.subplots()
bar_container = ax.bar(longitude, pop, width = 0.03)
ax.set(ylabel = 'Population', xlabel = 'Longitude (Bin size = 0.1)')
ax.text(-122.5, 505000, s = 'Portland', ha = 'center')
ax.text(-123.4, 415000, s = 'Salem/Eugene', ha = 'center')
ax.text(-121.3, 110000, s = 'Bend', ha = 'center')
#ax.text(-120.4, 2111, s = 'hello')

#plt.bar(-124.4, 21111, align='edge', width=0.03)

plt.rcParams["figure.figsize"] = (10,5.5)

plt.show()
```
