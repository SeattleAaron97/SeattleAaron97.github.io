```python
##### In the cell, enter the ELO of both players and get back the percent chance of each winning

def eloCalc(p1Elo, p2Elo):
    eloDiff = p1Elo - p2Elo
    #print(type(eloDiff))
    200 / 400
    result = 1 - (1 / (1 + (10 ** ((eloDiff) / 400))))
    return 'Player 1 has a ', result * 100, ' percent chance of winning according to the ELOs'

print(eloCalc(2203, 1947))


```

```python
#### +200 means said player has a %33 Chance of winning, while -200 means said player has a 66% chance of winning

### This function is for calculating the favorite (the '-' odds) 
def oddsWinPct(plus_minus, lineNum):
    if(plus_minus == '-'):
        result = ((lineNum / 100) / 1) / (lineNum + 100) * 100
    elif(plus_minus == '+'):
        result = 1 - ((lineNum / 100) / 1) / (lineNum + 100) * 100
    
    return ('according to the oddsmakers, player has a', result * 100, 'percent chance of winning')
    #return result    
    

#def calcUnderdogPct(lineNum):
    

#2/1 / (200 + 100) * 100

print(oddsWinPct('-', 145))
```
