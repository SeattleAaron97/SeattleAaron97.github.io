```python
import pandas as pd
import numpy as np
import matplotlib
from sklearn.tree import DecisionTreeClassifier

pd.set_option('display.max_columns', None) ### This line makes all the columns display, rather than ellipses shorten

atp_2022_frame = pd.read_csv('C:/Users/aaron/Code/Data/atp_matches_2022.csv')
```

```python
### Created 'upset ratio'. maybe not the best name, but the smaller the number(decimal), means the heavy favorite one.
### A larger number means it was a larger upset (by rank)
atp_2022_frame['upset_ratio'] = atp_2022_frame['winner_rank'] / atp_2022_frame['loser_rank']
```

```python
##### Step one. Make a simple table with P1 ranking, P2 ranking, and winner. Thats it. Try to run a Model on that.


simple_frame = pd.DataFrame()
simple_frame['Player_One_ID'] = atp_2022_frame['winner_id']
simple_frame['Player_One_Rank'] = atp_2022_frame['winner_rank']
simple_frame['Player_Two_ID'] = atp_2022_frame['loser_id']
simple_frame['Player_Two_Rank'] = atp_2022_frame['loser_rank']
simple_frame['Winner_Name'] = atp_2022_frame['winner_name']
```

```python
#### Drop rows where one of the players has no ranking

#bool1 = atp_2022_frame['winner_rank'].notnull()
#bool2 = atp_2022_frame['loser_rank'].notnull()

#ranking_mask = atp_2022_frame['winner_rank'].notnull() or atp_2022_frame['loser_rank'].notnull()
ranking_mask = atp_2022_frame['winner_rank'].notnull() & atp_2022_frame['loser_rank'].notnull()

ranking_mask.value_counts()

simple_frame = simple_frame[ranking_mask]

```
