#### Hello!! In this IPYNB file I will be attempting to clean and make some sense of the chicago data! here goes nothing :)

```python
import pandas as pd

#import the data!!!!
metro_data = pd.read_csv('Data/CTA_-_Ridership_-__L__Station_Entries_-_Daily_Totals.csv')

###
# various notes on data -
# - W === weekday, A === Saturday, U === sunday/holiday
#
# data source can be found here -> https://data.cityofchicago.org/Transportation/CTA-Ridership-L-Station-Entries-Daily-Totals/5neh-572f
# column names (all 5 of em!)---> station_id	stationname	date	daytype	rides
#
#metro_data.head()
```

```python
metro_data.describe()

metro_data.shape
```

```python
day_stats = metro_data.groupby('daytype').agg({'rides':'count'}).reset_index()

day_stats
```

#### I am going to find the most used stations on saturdays vs weekdays and see if it matters!

```python
only_weekdays = metro_data[metro_data['daytype'] == 'W']

only_weekdays = only_weekdays.sort_values('rides', ascending=False)

only_weekdays.head(10)
```

```python
only_saturdays = metro_data[metro_data['daytype'] == 'A']

only_saturdays = only_saturdays.sort_values('rides', ascending=False)

only_saturdays.head(10)
```

```python
only_sundays = metro_data[metro_data['daytype'] == 'U']

only_sundays = only_sundays.sort_values('rides', ascending=False)

only_sundays.head(10)
```

#### The above data is good, but not super informative because it only considers one day at a time. The next cells will describe stations over more extended time than just one day.

```python
#counts the number of unique values in the stationname column
unique_stations = metro_data.stationname.nunique()

unique_stations
```

```python
#attemping to find most used stations total

#successfully finding total rides for one station, not the best data though
arranged = metro_data.groupby('stationname')

central_lake = arranged.get_group('Central-Lake')
central_lake_ride_sum = central_lake['rides'].sum()

lake_state = arranged.get_group('Lake/State')
lake_state_ride_sum = lake_state['rides'].sum()

print(central_lake_ride_sum)
print(lake_state_ride_sum)
```

```python
#creates list of all stations with the corresponding number of rides
total_rides_by_station = metro_data.groupby('stationname').agg({'rides':'sum'}).reset_index()

total_rides_by_station = total_rides_by_station.sort_values('rides', ascending=False)

total_rides_by_station
```

```python
total_rides_by_station_only_weekdays = only_weekdays.groupby('stationname').agg({'rides':'sum'}).reset_index()

total_rides_by_station_only_weekdays = total_rides_by_station_only_weekdays.sort_values('rides', ascending=False)

total_rides_by_station_only_weekdays.head(10)
```

```python
total_rides_by_station_only_saturdays = only_saturdays.groupby('stationname').agg({'rides':'sum'}).reset_index()

total_rides_by_station_only_saturdays = total_rides_by_station_only_saturdays.sort_values('rides', ascending=False)

total_rides_by_station_only_saturdays.head(10)
```

```python
total_rides_by_station_only_sundays = only_sundays.groupby('stationname').agg({'rides':'sum'}).reset_index()

total_rides_by_station_only_sundays = total_rides_by_station_only_sundays.sort_values('rides', ascending=False)

total_rides_by_station_only_sundays.head(10)
```
