#### Ok now I am doing an analysis of deaths on I-35 vs Austin in general

```python
#### Calculate the percent of deaths, injuries and incidents that happen on 35 and in total

IH_35_deaths = on_35['death_cnt'].sum()
total_deaths = crash_data_frame['death_cnt'].sum()

#IH_35_deaths / total_deaths  ## = 192 / 885 = 0.21694915254237288

IH_35_injuries = on_35['tot_injry_cnt'].sum()
total_injuries = crash_data_frame['tot_injry_cnt'].sum()

#IH_35_injuries / total_injuries ## = 16543.0 / 95697.0 =  0.17286853297386542

IH_35_incidents = on_35['crash_id'].count()
total_incidents = crash_data_frame['crash_id'].count()

#IH_35_incidents / total_incidents # 25263 / 148579 = 0.17003075804790718
```

```python
I35_deaths = on_35.groupby('crash_year').agg({'death_cnt':'sum'})

I35_deaths
```

```python
total_deaths_by_year = crash_data_frame.groupby('crash_year').agg({'death_cnt':'sum'})

total_deaths_by_year
```

```python
(I35_deaths/total_deaths_by_year).plot(kind='bar',xlabel='Year',ylabel='Percent of Austin traffic deaths on I-35', title='I-35 Deaths by percent')
```

```python
reset_total = total_deaths_by_year.reset_index()
reset_35 = I35_deaths.reset_index()
#reset_35
```

```python
total_and_35_deaths = pd.DataFrame({'Total traffic deaths': reset_total['death_cnt'].tolist(), 'Traffic Deaths on IH-35': reset_35['death_cnt'].tolist()}, index=reset_total['crash_year'])

ax = total_and_35_deaths.plot.bar()
```

#### A quick look into serious injuries

```python
#Crash Severity - Most severe injury suffered by any one person involved in the crash
#( 0=UNKNOWN, 1=INCAPACITATING INJURY, 2=NON-INCAPACITATING INJURY, 3=POSSIBLE INJURY, 4=KILLED, 5=NOT INJURED)

serious_injury_mask = crash_data_frame['crash_sev_id'] == 1

serious_injury_frame = crash_data_frame[serious_injury_mask]

serious_injury_frame.head()
```

```python
serious_injury_by_year = crash_data_frame.groupby('crash_year').agg({'crash_sev_id':'count'})

serious_injury_by_year
```

```python
serious_injury_by_year_on_35 = on_35.groupby('crash_year').agg({'crash_sev_id':'count'})

serious_injury_by_year_on_35
```

```python
reset_total_serious_injuries = serious_injury_by_year.reset_index()
reset_35_serious_injuries = serious_injury_by_year_on_35.reset_index()

reset_35_serious_injuries
```

```python
total_and_35_serious_injuries = pd.DataFrame({'Total serious injuries': reset_total_serious_injuries['crash_sev_id'].tolist(), 'Serious injuries on IH-35': reset_35_serious_injuries['crash_sev_id'].tolist()}, index=reset_total_serious_injuries['crash_year'])

serious_injury_plot = total_and_35_serious_injuries.plot.bar()
```
