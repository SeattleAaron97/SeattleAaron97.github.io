```python
def calculate_route_stuff(specific_route_id):
    ### route id must be a string
    #specific_route = feed.stop_times[feed.stop_times.trip_id == 'LLR_2021-06-12-28_Saturday_1003']
    specific_route = feed.stop_times[feed.stop_times.trip_id == specific_route_id]

    specific_route['arrival_time'] = pd.to_datetime(specific_route['arrival_time'])
    specific_route['departure_time'] = pd.to_datetime(specific_route['departure_time'])
    first_stop = specific_route['stop_sequence'].min()
    last_stop = specific_route['stop_sequence'].max()

    number_of_stops = specific_route['stop_sequence'].nunique()

    first_stop_time = specific_route[specific_route['stop_sequence'] == first_stop]['arrival_time'].iloc[0]
    last_stop_time = specific_route[specific_route['stop_sequence'] == last_stop]['arrival_time'].iloc[0]

    total_trip_time = last_stop_time - first_stop_time

    return number_of_stops, total_trip_time
```

```python
calculate_route_stuff('LLR_2021-06-12-28_Saturday_1015')
```
